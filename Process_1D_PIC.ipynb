```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pickle

from pic.functions import smooth
```

```python
data = pickle.load(open("data.dat","rb"))
```

```python
def meanfield(key,data=data):
    ks = [k for k in data.keys()]
    
    vect = data[ks[0]][key]
    for k in ks[1:]:
        vect += data[k][key]
    vect /= len(ks)
    
    return vect[:-2]
```

```python
for key in ['ne','ni','Te','ve',"phi"]:
    vect = meanfield(key)
    plt.plot(vect, label = key)
    plt.legend()
    plt.show()
```

```python
Te = meanfield("Te")
phi = meanfield("phi")
ne = meanfield("ne")


plt.figure(figsize=(10,3.5))
plt.subplot(121)
plt.plot(phi)
plt.plot(Te)
plt.subplot(122)
plt.plot(Te/(phi[1:]))
plt.tight_layout()
plt.show()

nT = ne[:-1]*Te
plt.plot(smooth(ne*np.gradient(phi)))
plt.plot(smooth(np.gradient(nT)))
```

```python
ve = meanfield("ve")

plt.plot(smooth(ve*ne[:-1]))
```
